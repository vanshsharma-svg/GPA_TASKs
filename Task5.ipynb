Import the necessary Python libraries for data manipulation,

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from zoneinfo import ZoneInfo
```

Load the Google Play Store dataset and the User Reviews dataset into

Understand the structure of both datasets by checking their dimensions,

=====================================
# Step 4: Clean the Play Store Dataset
=====================================

Clean the Play Store dataset by removing duplicate records,
handling missing values, and converting important columns
to appropriate data types.

```python
# Remove duplicate apps
play_store_data = play_store_data.drop_duplicates(subset='App')

# Convert Rating
play_store_data['Rating'] = pd.to_numeric(
    play_store_data['Rating'],
    errors='coerce'
)

# Convert Reviews
play_store_data['Reviews'] = pd.to_numeric(
    play_store_data['Reviews'],
    errors='coerce'
)

# Clean Installs
play_store_data['Installs'] = (
    play_store_data['Installs']
    .astype(str)
    .str.replace(',', '', regex=False)
    .str.replace('+', '', regex=False)
)

play_store_data['Installs'] = pd.to_numeric(
    play_store_data['Installs'],
    errors='coerce'
)

# Clean Price
play_store_data['Price'] = (
    play_store_data['Price']
    .astype(str)
    .str.replace('$', '', regex=False)
)

play_store_data['Price'] = pd.to_numeric(
    play_store_data['Price'],
    errors='coerce'
)
```

=====================================
# Step 5: Convert App Size into MB
=====================================

Convert the Size column into numerical values (MB)
to make it suitable for visualization and analysis.

```python
def convert_size(size):

    if pd.isna(size):
        return np.nan

    size = str(size)

    if size.endswith("M"):
        return float(size[:-1])

    elif size.endswith("k"):
        return float(size[:-1]) / 1024

    else:
        return np.nan


play_store_data["Size_MB"] = play_store_data["Size"].apply(convert_size)

play_store_data = play_store_data.dropna(
    subset=["Rating", "Reviews", "Installs", "Size_MB"]
)

play_store_data.head()
```

=====================================
# Step 6: Clean the User Reviews Dataset
=====================================

Clean the User Reviews dataset by converting the
Sentiment Subjectivity column into numeric format,
removing missing values, and preparing the data
for merging.

```python
# Keep only the required columns
review = review[['App', 'Sentiment_Subjectivity']]

# Convert Sentiment_Subjectivity to numeric
review['Sentiment_Subjectivity'] = pd.to_numeric(
    review['Sentiment_Subjectivity'],
    errors='coerce'
)

# Remove missing values
review = review.dropna(subset=['Sentiment_Subjectivity'])

# If an app has multiple reviews, calculate the average subjectivity
review = review.groupby('App', as_index=False)['Sentiment_Subjectivity'].mean()

print("User Reviews Dataset Shape:", review.shape)

review.head()
```

=====================================
# Step 7: Merge the Datasets
=====================================

Merge the Play Store dataset with the User Reviews
dataset using the App column. This allows us to
combine app information with sentiment subjectivity.

```python
merged_data = pd.merge(
    play_store_data,
    review,
    on='App',
    how='inner'
)

print("Merged Dataset Shape:", merged_data.shape)

merged_data.head()
```

=====================================
# Step 8: Apply Business Requirements
=====================================

Filter the merged dataset according to the given
business requirements.

Conditions:

• Rating > 3.5

• Reviews > 500

• Installs > 50,000

• Sentiment Subjectivity > 0.5

• App name should not contain the letter 'S'

• Categories:
  - Game
  - Beauty
  - Business
  - Comics
  - Communication
  - Dating
  - Entertainment
  - Social
  - Events

```python
# Required Categories
required_categories = [
    'GAME',
    'BEAUTY',
    'BUSINESS',
    'COMICS',
    'COMMUNICATION',
    'DATING',
    'ENTERTAINMENT',
    'SOCIAL',
    'EVENTS'
]

# Apply all filters
filtered_data = merged_data[
    (merged_data['Rating'] > 3.5) &
    (merged_data['Reviews'] > 500) &
    (merged_data['Installs'] > 50000) &
    (merged_data['Sentiment_Subjectivity'] > 0.5) &
    (~merged_data['App'].str.contains('S', case=False, na=False)) &
    (merged_data['Category'].isin(required_categories))
].copy()

print("Filtered Dataset Shape:", filtered_data.shape)

filtered_data.head()
```

```python
print(sorted(merged_data['Category'].unique()))
```

=====================================
# Step 9: Translate Category Names
=====================================

Translate selected category names for visualization purposes only.

Translations:

• Beauty → सौंदर्य (Hindi)

• Business → வணிகம் (Tamil)

• Dating → Partnersuche (German)

The original dataset remains unchanged.

```python
# Translate category names only for display
category_translation = {
    'BEAUTY': 'सौंदर्य',
    'BUSINESS': 'வணிகம்',
    'DATING': 'Partnersuche'
}

filtered_data['Category_Display'] = (
    filtered_data['Category']
    .replace(category_translation)
)

filtered_data[['Category', 'Category_Display']].drop_duplicates()
```

=====================================
# Step 10: Check Time Constraint (IST)
=====================================

Verify the current Indian Standard Time (IST).

The Bubble Chart will be displayed only if the
current time is between **5:00 PM IST** and **7:00 PM IST**.
Otherwise, a message will be displayed.

```python
# Get current IST time
current_time = datetime.now(ZoneInfo("Asia/Kolkata")).time()

# Define allowed time
start_time = datetime.strptime("17:00", "%H:%M").time()
end_time = datetime.strptime("19:00", "%H:%M").time()

print("Current IST Time :", current_time)
```

=====================================
# Step 11: Plot the Bubble Chart
=====================================

Create a Bubble Chart to visualize the relationship
between App Size (MB) and App Rating.

Chart Details:

• X-axis → App Size (MB)

• Y-axis → App Rating

• Bubble Size → Number of Installs

• Bubble Color → Category

• Game Category highlighted in Pink

• Beauty, Business and Dating displayed in translated languages.

```python
if start_time <= current_time <= end_time:

    plt.figure(figsize=(15,8))

    # Category colors
    colors = {
        'GAME': 'pink',
        'सौंदर्य': 'gold',
        'வணிகம்': 'green',
        'COMICS': 'orange',
        'COMMUNICATION': 'blue',
        'Partnersuche': 'red',
        'ENTERTAINMENT': 'purple',
        'SOCIAL': 'cyan',
        'EVENTS': 'gray'
    }

    for category in filtered_data['Category_Display'].unique():

        temp = filtered_data[
            filtered_data['Category_Display'] == category
        ]

        original_category = temp['Category'].iloc[0]

        if original_category == "GAME":
            color = "pink"
        else:
            color = colors.get(category, "black")

        plt.scatter(
            x=temp["Size_MB"],
            y=temp["Rating"],
            s=temp["Installs"]/1000,
            c=color,
            alpha=0.7,
            edgecolors="black",
            linewidth=0.5,
            label=category
        )

    plt.title(
        "Bubble Chart: App Size vs Rating",
        fontsize=16,
        fontweight="bold"
    )

    plt.xlabel("App Size (MB)", fontsize=12)

    plt.ylabel("Average Rating", fontsize=12)

    plt.grid(True, linestyle="--", alpha=0.5)

    plt.legend(title="Category")

    plt.tight_layout()

    plt.show()

else:

    print("Bubble chart is available only between 5 PM IST and 7 PM IST.")
```
