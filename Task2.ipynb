# =====================================
# Step 1: Import Libraries
# =====================================

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from datetime import datetime
from zoneinfo import ZoneInfo
```

# =====================================
# Step 2: Load data from csv
# =====================================

```python
play_store_data = pd.read_csv(
    r"C:\Users\Vansh Sharma\Downloads\Play Store Data (1).csv"
)

play_store_data.head()
```

# =====================================
# Step 3: Clean the Installs Column
# =====================================

```python
play_store_data["Installs"] = (
    play_store_data["Installs"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace("+", "", regex=False)
)

play_store_data = play_store_data[
    play_store_data["Installs"] != "Free"
]

play_store_data["Installs"] = play_store_data["Installs"].astype(int)
```

# =====================================
# Step 4: Clean Price Column
# =====================================

```python
play_store_data["Price"] = (
    play_store_data["Price"]
    .astype(str)
    .str.replace("$", "", regex=False)
)

play_store_data["Price"] = play_store_data["Price"].astype(float)
play_store_data["Price"].head()
```

# =====================================
# Step 5: Create Revenue Column
# =====================================

```python
play_store_data["Revenue"] = (
    play_store_data["Price"] *
    play_store_data["Installs"]
)
play_store_data[
    ["App","Type","Price","Installs","Revenue"]
].head()
```

# =====================================
# Step 6: Clean Android Version Column
# =====================================

```python
play_store_data["Android Version"] = (
    play_store_data["Android Ver"]
    .str.extract(r'(\d+\.\d+)')
)

play_store_data["Android Version"] = pd.to_numeric(
    play_store_data["Android Version"],
    errors="coerce"
)
```

```python
play_store_data[["Android Ver", "Android Version"]].head()
```

# =====================================
# Step 7: Clean Size Column (Convert to MB)
# =====================================

```python
play_store_data["Size_MB"] = (
    play_store_data["Size"]
    .str.replace("M", "", regex=False)
)

play_store_data["Size_MB"] = pd.to_numeric(
    play_store_data["Size_MB"],
    errors="coerce"
)
```

```python
play_store_data[["Size", "Size_MB"]].head()
```

# =====================================
# Step 8: Calculate App Name Length
# =====================================

```python
play_store_data["App_Length"] = (
    play_store_data["App"].str.len()
)
play_store_data[["App", "App_Length"]].head()
```

# =====================================
# Step 9: Apply Filtering Conditions
# =====================================

```python
filtered_data = play_store_data[
    (play_store_data["Installs"] >= 10000) &
    (play_store_data["Revenue"] >= 10000) &
    (play_store_data["Android Version"] > 4.0) &
    (play_store_data["Size_MB"] > 15) &
    (play_store_data["Content Rating"] == "Everyone") &
    (play_store_data["App_Length"] <= 30)
]

print(filtered_data.shape)
filtered_data.head()
```

# =====================================
# Step 10: Get Top 3 Categories
# =====================================

```python
top3_categories = (
    filtered_data.groupby("Category")["Installs"]
    .sum()
    .nlargest(3)
    .index
)

top3_data = filtered_data[
    filtered_data["Category"].isin(top3_categories)
]
print(top3_categories)
top3_data.head()
```

# =====================================
# Step 11: Aggregate Data (Avg Installs & Revenue)
# =====================================

```python
agg_data = (
    top3_data.groupby(["Category", "Type"])
    .agg(
        Avg_Installs=("Installs", "mean"),
        Avg_Revenue=("Revenue", "mean")
    )
    .reset_index()
)

agg_data
```

# =====================================
# Step 12: Create Dual-Axis Plot (Plotly)
# =====================================

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(specs=[[{"secondary_y": True}]])

colors = {
    "Free": "#1f77b4",
    "Paid": "#ff7f0e"
}

for app_type in agg_data["Type"].unique():
    df_type = agg_data[agg_data["Type"] == app_type]

    # 📊 Bars - Installs
    fig.add_trace(
        go.Bar(
            x=df_type["Category"],
            y=df_type["Avg_Installs"],
            name=f"{app_type} Avg Installs",
            marker_color=colors.get(app_type, "#636EFA"),
            text=df_type["Avg_Installs"].round(0),
            textposition="outside"
        ),
        secondary_y=False
    )

    # 📈 Line - Revenue
    fig.add_trace(
        go.Scatter(
            x=df_type["Category"],
            y=df_type["Avg_Revenue"],
            name=f"{app_type} Avg Revenue",
            mode="lines+markers+text",
            line=dict(width=3),
            text=df_type["Avg_Revenue"].round(0),
            textposition="top center"
        ),
        secondary_y=True
    )

fig.update_layout(
    title="🔥 Top Categories: Avg Installs vs Revenue (Free vs Paid)",
    template="plotly_white",
    hovermode="x unified"
)

fig.update_yaxes(title_text="Average Installs", secondary_y=False)
fig.update_yaxes(title_text="Average Revenue ($)", secondary_y=True)

fig.show()
```

# =====================================
# Step 13: Time Condition (1 PM to 2 PM IST)
# =====================================

```python
ist = ZoneInfo("Asia/Kolkata")
current_time = datetime.now(ist).time()

start = datetime.strptime("13:00", "%H:%M").time()
end = datetime.strptime("14:00", "%H:%M").time()

if start <= current_time <= end:

    fig = make_subplots(specs=[[{"secondary_y": True}]])

    colors = {"Free": "#1f77b4", "Paid": "#ff7f0e"}

    for app_type in agg_data["Type"].unique():
        df_type = agg_data[agg_data["Type"] == app_type]

        fig.add_trace(
            go.Bar(
                x=df_type["Category"],
                y=df_type["Avg_Installs"],
                name=f"{app_type} Avg Installs",
                marker_color=colors.get(app_type),
            ),
            secondary_y=False
        )

        fig.add_trace(
            go.Scatter(
                x=df_type["Category"],
                y=df_type["Avg_Revenue"],
                name=f"{app_type} Avg Revenue",
                mode="lines+markers"
            ),
            secondary_y=True
        )

    fig.update_layout(
        title="🔥 Top Categories: Avg Installs vs Revenue",
        template="plotly_white",
        hovermode="x unified"
    )

    fig.update_yaxes(title_text="Average Installs", secondary_y=False)
    fig.update_yaxes(title_text="Average Revenue ($)", secondary_y=True)

    fig.show()

else:
    print("Graph is available only between 1 PM and 2 PM IST.")
print(current_time)
```
